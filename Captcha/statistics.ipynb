```python
import csv
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
csv_file = csv.reader(open('./data/captcha/labels/labels.csv', 'r'))
```

```python
lens = np.zeros([4])
for i in csv_file:
    x = len(i[1])
    if x == 1:
        lens[0] += 1 
    if x == 2:
        lens[1] += 1
    if x == 3:
        lens[2] += 1
    if x == 4:
        lens[3] += 1
print(lens)
images_num = lens[0] + lens[1] + lens[2] +lens[3]
print("images_num: %d" % images_num)
```

```python
len_labels = [1,2,3,4]
fig = plt.figure(1)
ax1 = plt.subplot(111)
data = np.array(lens)
width = 0.5
x_bar = np.arange(4)
rect = ax1.bar(left = x_bar,height = data, width = width,color = "blue")
for rec in rect:
    x = rec.get_x()
    height = rec.get_height()
    ax1.text(x+0.1, 1.02*height ,str(height))
ax1.set_xticks(x_bar)
ax1.set_xticklabels(tuple(len_labels))
ax1.set_ylabel("numbers")
ax1.set_title("kinds of Numbers")
ax1.grid(True)
ax1.set_ylim(0,18000)
plt.show()
```

```python
pros = np.zeros([4])
for i in range(4):
    pros[i]=lens[i]/images_num
plt.axes(aspect = 1)
plt.pie(x = pros, labels = len_labels, autopct = '%3.1f %%', 
       shadow = True, labeldistance = 1.1, startangle = 90, pctdistance = 0.6)
plt.show()
```
